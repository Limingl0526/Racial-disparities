```python
import math
import matplotlib.pyplot as plt


def okamura(fc, hb, r):

    a = 69.55 + 26.16*math.log10(fc) - 13.82*math.log10(hb)

    b = 44.9 - 6.55*math.log10(hb)

    c = 5.4 + 2*(math.log10(fc/28))**2

    d = 40.94 + 4.78*(math.log10(fc))**2 - 18.33*math.log10(fc)


    L_urbana = a + b*math.log10(r)
    L_suburbana = a + b*math.log10(r) -c
    L_aberta = a + b*math.log10(r) -d

    return L_urbana, L_suburbana, L_aberta

fc = 150
hb = 50
r = 2

l_fc1, l_fc2, l_fc3 = [], [], []
eixo_fc = []

#Variar fc
for i in range(150, 1500, 10):

    l1, l2, l3 = okamura(i, hb, r)

    l_fc1.append(l1)
    l_fc2.append(l2)
    l_fc3.append(l3)
    eixo_fc.append(i)


plt.plot(eixo_fc, l_fc1, label = "L Urban")
plt.plot(eixo_fc, l_fc2, label = "L Suburban")
plt.plot(eixo_fc, l_fc3, label = "L plain")

plt.title('fc: hb = 50m, R = 2Km')
plt.xlabel('fc (MHz)')
plt.ylabel('L_T (dB)')

plt.legend()

plt.show()

#Variando hb
fc = 200
eixo_hb = []
l_hb1, l_hb2, l_hb3 = [], [], []

for i in range(30, 200, 5):

    l1, l2, l3 = okamura(fc, i, r)

    l_hb1.append(l1)
    l_hb2.append(l2)
    l_hb3.append(l3)
    eixo_hb.append(i)


plt.plot(eixo_hb, l_hb1, label = "L Urban")
plt.plot(eixo_hb, l_hb2, label = "L Suburban")
plt.plot(eixo_hb, l_hb3, label = "L plain")

plt.title(' fc = 200MHz, R = 2Km')
plt.xlabel('hb (m)')
plt.ylabel('L_T (dB)')

plt.legend()

plt.show()


#Variando raio
hb = 50
eixo_r = []
l_r1, l_r2, l_r3 = [], [], []

for i in range(1, 15, 1):

    l1, l2, l3 = okamura(fc, hb, i)

    l_r1.append(l1)
    l_r2.append(l2)
    l_r3.append(l3)
    eixo_r.append(i)


plt.plot(eixo_r, l_r1, label = "L Urban")
plt.plot(eixo_r, l_r2, label = "L Suburban")
plt.plot(eixo_r, l_r3, label = "L plain")

plt.title('fc = 200MHz, hb = 50m')
plt.xlabel('R (Km)')
plt.ylabel('L_T (dB)')

plt.legend()

plt.show()
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import shape

def main():
    # this is the name of the geography you want to retrieve. update to meet your needs
    location = 'Japan'

    dataset_links = pd.read_csv("https://minedbuildings.blob.core.windows.net/global-buildings/dataset-links.csv")
    greece_links = dataset_links[dataset_links.Location == location]
    for _, row in greece_links.iterrows():
        df = pd.read_json(row.Url, lines=True)
        df['geometry'] = df['geometry'].apply(shape)
        gdf = gpd.GeoDataFrame(df, crs=4326)
        gdf.to_file(f"{row.QuadKey}.geojson", driver="GeoJSON")


if __name__ == "__main__":
    main()
```

```python
import geopandas as gpd
import pandas as pd
import dask.dataframe as dd
import os

# 设定一个合理的批处理大小，这取决于文件大小和可用内存
batch_size = 10

for i in range(1, 101):
    folder_name = f"dest_folder_{i}"  # 补上 i 来生成不同的文件夹名
    folder_path = f"E:/2024.1.14/data_splite/{folder_name}"
    
    if not os.path.exists(folder_path):
        print(f"Folder does not exist: {folder_path}")
        continue
        
    geojson_files = [f for f in os.listdir(folder_path) if f.endswith('.geojson')]

    # 创建一个空的GeoDataFrame来存储最终的合并结果
    combined_gdf = gpd.GeoDataFrame()
    
    # 通过文件列表进行分批处理
    for batch_start in range(0, len(geojson_files), batch_size):
        gdf_list = []
        for filename in geojson_files[batch_start:batch_start + batch_size]:
            file_path = os.path.join(folder_path, filename)
            
            # 检查路径是否正确
            print(f"Checking file path: {file_path}")
            if not os.path.exists(file_path):
                print(f"File does not exist: {file_path}")
                continue
            
            # 尝试读取文件
            try:
                gdf = gpd.read_file(file_path)
            except Exception as e:
                print(f"Failed to read file {file_path}: {e}")
                continue
            
            gdf_list.append(gdf)
        
        if gdf_list:  # 如果当前批次非空，则进行合并
            # 合并当前批次内的GeoDataFrame对象
            batch_combined = pd.concat(gdf_list, ignore_index=True)
            del gdf_list  # 及时删除列表释放内存
            
            # 使用Dask来优化concat合并的性能
            dask_gdf = dd.from_pandas(batch_combined, npartitions=10)
            del batch_combined  # 删除中间变量以节省内存
            
            # 计算dask dataframe并转换为GeoDataFrame
            gdf_batch = gpd.GeoDataFrame(dask_gdf.compute())
            combined_gdf = pd.concat([combined_gdf, gdf_batch], ignore_index=True)
            del gdf_batch  # 删除中间变量以节省内存

    # 储存最终合并的文件
    if not combined_gdf.empty:
        output_filename = f'combined_{folder_name}.geojson'
        output_path = os.path.join(folder_path, output_filename)
        os.makedirs(folder_path, exist_ok=True)
        combined_gdf.to_file(output_path, driver='GeoJSON')
        print(f'All GeoJSON files in {folder_name} have been combined into {output_filename}')
    else:
        print(f'No GeoJSON files to combine in {folder_name}')
```

```python
import geopandas as gpd
import pandas as pd
import os

# Function to combine all GeoJSON files in a given folder
def combine_geojson_files(folder_path, output_folder, folder_name):
    geojson_files = [f for f in os.listdir(folder_path) if f.endswith('.geojson')]

    gdf_list = []
    for filename in geojson_files:
        file_path = os.path.join(folder_path, filename)
        gdf = gpd.read_file(file_path)
        gdf_list.append(gdf)

    combined_gdf = gpd.GeoDataFrame(pd.concat(gdf_list, ignore_index=True))

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
  
    output_filename = f'combined_{folder_name}.geojson'  # 指定合并后的文件名
    output_path = os.path.join(output_folder, output_filename)

    combined_gdf.to_file(output_path, driver='GeoJSON')

    print(f'All GeoJSON files in {folder_path} have been combined into {output_filename}')

# Base path where all the 'dest_folder_X' are located
base_path = "E:/2024.1.14/datasplite/"

# Output folder to store all combined GeoJSON files
output_base_folder = "E:/2024.1.14/output/"

# Perform the combination operation on each folder
for i in range(10, 12):
    dest_folder = f"dest_folder_{i}"  # 确保folder_name在这里是唯一的
    folder_path = os.path.join(base_path, dest_folder)

    # 这里我们不再使用 dest_folder_name 作为输出文件夹名称，因为我们想要所有文件在同一个地方
    combine_geojson_files(folder_path, output_base_folder, dest_folder)
    print('Completed combining GeoJSON')

print('Completed combining GeoJSON files in all folders.')
```
